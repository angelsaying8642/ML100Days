## 作業
在鐵達尼資料集中，
透過進階補值方法 KNN，將鐵達尼號中的 Titanic_train
.csv 中的 age 的遺失值進行補值。    
* step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性?
* step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
```

## 讀入資料

```python
data = pd.read_csv('Titanic_train.csv')
```

```python
data
```

```python
#計算每一行是否有遺失值，計算遺失比例
# 計算遺失值比例
missing_vals = data.isnull().sum()/len(data)
# 遺失值比例排序
missing_vals.sort_values(ascending=False)
# 建立 DataFrame
missing_vals = pd.DataFrame(missing_vals,columns=['missing_rate'])
print(missing_vals)
# age 有遺失值
```

## 觀察 Age

```python
# age 與 sex 的差異
## x 要擺類別行變數 y要擺連續型變數
g = sns.catplot(x="Sex", y="Age", hue="Sex",
               data=data,kind="box")
# 光看箱型圖沒看出有差異
```

```python
# age 與 sex 的差異
## x 要擺類別行變數 y要擺連續型變數
g = sns.catplot(x="Pclass", y="Age", hue="Pclass",
               data=data,kind="box")
# 看箱型圖有差異
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
```

```python
data
```

```python
data = data[['Sex', 'Age', 'Pclass']]
from sklearn.metrics.pairwise import nan_euclidean_distances
display(pd.DataFrame(nan_euclidean_distances(data)))
```

```python
# KNN
value_neighbors=2
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors,add_indicator= True)
df_filled = pd.DataFrame(imputer.fit_transform(data))
```
